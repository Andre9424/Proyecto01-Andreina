```python
import folium
import pandas as pd

# Impresión de la versión de Folium
print(folium.__version__)

# Impresión de la versión de Pandas
print(pd.__version__)
```

```python
#Iniciar Mapa
m = folium.Map(
    location=[10, -84],
    tiles='OpenStreetMap',
    zoom_start=7, 
    control_scale=True)
 
#Mapas Base  
folium.TileLayer(
    tiles='http://services.arcgisonline.com/arcgis/rest/services/World_Topo_Map/MapServer/MapServer/MapServer/tile/{z}/{y}/{x}', 
    attr='ESRI World Imagery', 
    name='Topografía').add_to(m)
folium.TileLayer(
    tiles='http://services.arcgisonline.com/arcgis/rest/services/World_Imagery/MapServer/MapServer/tile/{z}/{y}/{x}', 
    attr='ESRI World Imagery', 
    name='Satelital').add_to(m)
folium.TileLayer(
    tiles='Stamen Terrain', 
    name='Relieve').add_to(m)

#Capas sobrepuestas
#Casos positivos
df_positivos = pd.read_csv('DatosProyecto/10_07_CSV_POSITIVOS_ULTIMA_FECHA.csv', sep=',')
df_positivos.rename(columns = {'Unnamed: 1': 'id'}, inplace=True)
geo_cantones = r'DatosProyecto/cr_cantones_wgs84.geojson'
bins = list(df_positivos['positivos'].quantile([0, 0.70, 0.90, 0.95, 1]))
m.choropleth(
    geo_data=geo_cantones,
    name='Casos Positivos',
    data=df_positivos,
    columns=['cod_canton', 'positivos'],
    bins=bins,
    key_on='feature.properties.cod_canton',
    fill_color='Reds', 
    fill_opacity=0.7,
    line_opacity=0.5,
    legend_name='Casos Positivos',
    smooth_factor=0)

#Casos activos
df_activos = pd.read_csv('DatosProyecto/10_07_CSV_ACTIVOS_ULTIMA_FECHA.csv', sep=',')
df_activos.rename(columns = {'Unnamed: 1': 'id'}, inplace=True)
geo_cantones = r'DatosProyecto/cr_cantones_wgs84.geojson'
bins = list(df_activos['activos'].quantile([0, 0.70, 0.90, 0.95, 1]))
m.choropleth(
    geo_data=geo_cantones,
    name='Casos Activos',
    data=df_activos,
    columns=['cod_canton', 'activos'],
    bins=bins,
    key_on='feature.properties.cod_canton',
    fill_color='YlOrRd', 
    fill_opacity=0.5, 
    line_opacity=0.5,
    legend_name='Casos Activos',
    smooth_factor=0)

#Casos recuperados
df_recuperados = pd.read_csv('DatosProyecto/10_07_CSV_RECUPERADOS_ULTIMA_FECHA.csv', sep=',')
df_recuperados.rename(columns = {'Unnamed: 1': 'id'}, inplace=True)
geo_cantones = r'DatosProyecto/cr_cantones_wgs84.geojson'
bins = list(df_recuperados['recuperados'].quantile([0, 0.70, 0.90, 0.95, 1]))
m.choropleth(
    geo_data=geo_cantones,
    name='Casos Recuperados',
    data=df_recuperados,
    columns=['cod_canton', 'recuperados'],
    bins=bins,
    key_on='feature.properties.cod_canton',
    fill_color='RdYlGn', 
    fill_opacity=0.5, 
    line_opacity=0.5,
    legend_name='Casos Recuperados',
    smooth_factor=0)

#Casos fallecidos
df_fallecidos = pd.read_csv('DatosProyecto/10_07_CSV_FALLECIDOS_ULTIMA_FECHA.csv', sep=',')
df_fallecidos.rename(columns = {'Unnamed: 1': 'id'}, inplace=True)
geo_cantones = r'DatosProyecto/cr_cantones_wgs84.geojson'
bins = list(df_fallecidos['fallecidos'].quantile([0, 0.45, 0.75, 0.95, 1]))
m.choropleth(
    geo_data=geo_cantones,
    name='Casos Fallecidos',
    data=df_fallecidos,
    columns=['cod_canton', 'fallecidos'],
    bins=bins,
    key_on='feature.properties.cod_canton',
    fill_color='YlGnBu', 
    fill_opacity=0.7,
    line_opacity=0.5,
    legend_name='Casos Fallecidos',
    smooth_factor=0)

folium.LayerControl().add_to(m)
m.save('Proyecto01_Andreina.html')
m
```
